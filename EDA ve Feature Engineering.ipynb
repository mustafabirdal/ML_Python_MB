# Feature Engineering

* Feature Engineerin veri setini modellemeye hazır hale getirmek ve veri setinin anlamını daha da güçlendirmek için uygulanan teknikleri içermektedir.  

**Feature Engineering'in ilk aşaması ve en önemli aşaması Data Splitting olacaktır. Bu aşamada veri seti Train, Validation ve Test seti olmak üzere 3 ayrı gruba ayrılacaktır.**

* Feature Engineering'deki tüm işlemler önce Train setine uygulanmak zorundadır. Daha sonra Train'de nasıl uygulanmışsa aynı şekilde(her şeyiyle aynı şekilde) Validation ve Test setlerine de uygulanmak zorundadır.  


**Feature Engineering Adımları**  
1.Data Splitting

## Data Splitting

* Hem ML modelimizin eğitilebilmesi hem overfitting olayından kaçınmak hem de data leakage denilen problemin önüne geçebilmek adına veri setimizi 3 ayrı parçaya böleriz. Bu parçalara Train, Test ve Validation setleri denilir.  


* **Train:** ML modelinin eğitilmesinde kullanılır. Feature Engineerin aşamaları ilk olarak bu set üzerinde kullanılır ve bu sette nasıl kullanılmışlarsa diğerlerinde de aynı şekilde kullanılırlar.  


* **Validation:** Yalnızca ML modelinin hyperparameter tuning aşamasında kullanılan veri setidir. ML modeli eğitilirken bu veriyi hiç görmez.  


* **Test:** Oluşturulmuş ML modelinin performansını test etmek amacıyla kullanılan veri setidir. ML modeli eğitilirken ve tuning yapılırken bu veriyi hiç görmez.  


* Biz Feature Engineering'in ilk önce Train'de uygulanmasını istediğimiz için Data Splitting adımını en başta yapmak istiyoruz. Böylece data leakage ve overfitting gibi problemlerin önlemiş oluruz.  


* Veri setinin bölünmesi %60(Train)-%20(Validation)-%20(Test) olacak şekilde ayarlanmalıdır.  


* Farklı durumlarda farklı yüzdelik dilimlerde de ayırabilirsin ama asla unutma ki Train setinin toplam veri setine göre yüzdesi %50'den aşağıda OLAMAZ!!!  


* Split aşamasında bir setin içindeki değerin diğerlerinde olmaması problem çıkarır. Buna dikkat etmek gerekir. Değerlerin homojen şekilde tüm setlere dağılmış olması istenir. Yani Test'de "A" diye bir değer varken diğerlerinde bu değer yoksa bu bir problem oluşturur.  


* <span class="mark">***Model kurarken farklı Train/Test/Validation oranlarıyla çalış, en iyi olanı seç.***</span>


* <span class="mark">***Ayrıca splitten sonra bağımsız kategorik değişkenlerdeki sınıfların nasıl dağıtıldığına da bak. Birinde olan diğerinde olmayabilir belki, bu durum encoding kısmında dikkat edilmesi gereken bir husus olabilir.***</span>   


### Yöntemler - Data Splitting
* Biz burada 2 tür splitting tekniği göstereceğiz. Normal yöntem ve Stratified yöntem.  


* Normal yöntemde datayı sadece belirtilen yüzdelik oranlarla böleceğiz. Bu yöntem hem classification hem de regression problemlerinde kullanılabilir.  


* Stratified yöntem ***imbalanced verisi olan classification problemlerinde*** kullanılır. Veriyi bölerken Output değişkenindeki sınıfların kendi içlerindeki yüzdeleri korunur.

#### Normal Train_Test_Validation Split

```python
# kullanılacak kütüphane
from sklearn.model_selection import train_test_split
df = pd.read_csv("train.csv")
df1 = df.copy()
```

```python
# Veriseti X=bağımsız değişkenler ve Y=output olarak ayrılır.
X = df.drop(["Survived"],axis=1)
y = df.Survived
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True, random_state=2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle=True, random_state=2)
```

```python
# Shuffle özelliğini açtığımız için index'ler karışmış halde olacaktır. Indexleri düzene sokmak için şu kodları yaz;
for i in [X_train, X_val, X_test, y_train, y_val, y_test]:
    i.reset_index(inplace=True, drop=True)
```

```python
print("X_train: ",X_train.shape," ","y_train: ",y_train.shape)
print("X_val: ",X_val.shape," ","y_val: ",y_val.shape)
print("X_test: ",X_test.shape," ","y_test: ",y_test.shape)
```

#### Stratified Train_Test_Validation Split

* bu teknikte diğerinden farklı olarak "stratify=" argümanını kullanırız ve bunu output değişkenine eşitleriz.

```python
df = pd.read_csv("diamonds.csv")
df1 = df.copy()
```

```python
# Data Split öncesi output değişkenindeki sınıfların frekans dağılımları.
df.price.value_counts(normalize=True)
```

```python
# Veriseti X=bağımsız değişkenler ve Y=output olarak ayrılır.
X = df.drop(["price"],axis=1)
y = df.price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True, stratify=y, random_state=2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle=True, stratify=y_train,
                                                  random_state=2)
```

## Handling Missing Values

* Eksik veriler veri kaybına sebep olur bu nedenle uygun tekniklerle doldurulmaları tavsiye edilir.  

* Bir çok algoritma eksik verinin varlığında çalışmaz.  

* Veri Biliminde 3 tip eksik veriden bahsedilir.  
1. **MCAR =** Missing Completely At Random. Yani veri eksikliği tamamen rastlantısal. Örn; birisi bir bilgiyi girmeyi unutmuş.  
2. **MAR =** Missing At Random. Verinin toplanış şekli nedeniyle oluşmuş bir eksiklik söz konusudur. Örn; Titanic verisinde veri kazadan sonra toplanmış ve bu nedenle bir çok vefat eden kazazedenin yaş bilgisine ulaşılamamış.  
3. **MNAR =** Missing Not At Random. Yani eksiklik hem verinin toplanması hem de verisetindeki bir başka değişkenle bağlantılıdır. Örn; tansiyon hastası olmayan bir kişi sık sık tansiyonunu ölçmeyeceği için bu ölçüm değerleri girilmemiş olabilir. Veya bir değişkende eksik veri varken aynı gözlemde başka bir değişkende de eksiklik varsa burada bir ilişki olabilir.  

**Kaynak;  
https://www.analyticsvidhya.com/blog/2020/07/knnimputer-a-robust-way-to-impute-missing-values-using-scikit-learn/
### Yapılacaklar
**1.Missing Value'ları bul, görüntüle ve görselleştir. (isnull, info gibi kodları kullan)**  

2.Missing Value'ların random mu yoksa ilişkisel bir sebepten mi oluştuğunu irdele. Örneğin kredi kartı numaralarını gösteren bir değişkende kredi kartı olmayanlar için değerler Nan olacaktır. Bu, ilişkisel bir sebeple oluşmuş bir missing datadır. Ancak bir veride hiçbir anlamı yokken doğum tarihi bilgisi eksikse bu da bir random missing value olacaktır. 


**3.Eğer değişkenin %50'si filan eksikse ve eksik veriler random sebeplerden oluşmuşsa o değişkeni komple silmek daha iyi olacaktır.**  

4.Eksik verinin durumuna göre uygulanacak yöntemi seç. Ayrıca kullandığın verisetinin bir kopyasını alarak (copy() fonksiyonu ile) uygulanan teknik sonrası değişkenin son halinin ve ilk halinin ne kadar değiştiğini std() fonksiyonu ile gör. ***Son bahsedilen işlem sadece continous datalar ile ilgilidir.***


5.**Missing Value'lara müdahale ederken önemli olan şey; uygulanan teknikten sonra verinin anlamının(yani korelasyon) ve dağılımının(distribution) değişmemesidir. Ufak değişimler göz ardı edilebilir.**

### Uygulanacak Yöntemler - Missing Values

**1. Silmek**; Bunu çok tercih etmesek de 1 milyon veriniz varken 10-15 tane eksik veriyi silmek bir problem oluşturmayacaktır.***(Eksik veri çok az sayıdaysa veya değişkenin %50'sinden fazlası random şekilde oluşmuş eksik verilerse bu yöntemi kullanıyorum)***      

**2. Eksik Veri için Değişken Yeni Oluşturma;** Eksik veri eğer ilişkisel bir eksik veriyse söz konusu değişkendeki eksik verilerin varlığını "0: eksik veri yok" ve "1: eksik veri var" şeklinde gösteren yeni bir değişken oluşturulur. Bu işlemden sonra gerekirse eksik verinin araştırıldığı bu değişken silinerek veriden atılabilir.(yani yukarıda anlatıldığı gibi eksik verinin rastgele değil veriyle ilgili ilişkisel bir sebepten oluşmuş olmasından bahsedilmektedir)***(İlişkisel eksik veri varsa kullanırım)***    

**3. Continous verilerde Median ortalaması ile eksik verileri doldurabilirsin;** Mean yerine Media seçilmesinin sebebi Normal Dağılımlı olmayan verilerde Mean değerinin outlier'lardan etkilenecek olmasıdır. Ayrıca değişkendeki tüm değerleri Median ile doldurmak veride yanlılığa yol açabilir. Bunu engellemek için eksik verisi olan değişkenin Median'ını almaktansa bu değişkeni başka bir kategorik değişkene göre gruplara ayırıp her bir gruba denk gelen Median ortalamasını eksik değerlere atarız. Örneğin; öğrencilerin yaşını veren bir değişkende eksik verilerimiz var diyelim. Öğrencilerimizi ortaokul, lise, üniversite gibi sınıfları bulunan bir kategorik değişkene göre gruplarsın. Her bir sınıfa denk gelen median yaş ortalamasını da öğrencilerin yaşlarını veren değişkende öğrencilerin sınıflarına göre eksik verilere atarsın.***(Eksik veri az sayıdaysa kullanıyorum)*** 

**4. Kategorik verilerde Mode ile eksik verileri doldurabilirsin;** Bu yöntem veride bir yanlılığa sebep olabilir. Bunu engellemek için eksik verisi olan değişkeninizi verisetindeki herhangi bir kategorik değişkene göre gruplarsın ve her bir gruba göre Mode'u alıp eksik verilerine o değeri atarsın.***(Eksik veri az sayıdaysa kullanıyorum)***   

**5. Rastgele Değerler ile Doldurmak;** Değişkendeki eksik veriler yine değişkende bulunan diğer verilerden rasgele seçilerek doldurulur.***(Eksik veri az sayıdaysa kullanıyorum)***   

**6. Algoritmalardan yararlanılabilir;** Eksik verinin bulunduğu kolon "Output" olarak düşünülür. Geri kalan diğer bağımsız değişkenler de yine bağımsız değişken olarak kalır. Bilinen değerler train datası olarak kullanılır ve istenilen bir algoritmanın kullanılmasıyla missing value'lar tahmin edilerek doldurulur.***(Eksik veri fazlaysa kullanıyorum)***  

**7. Kategorik verilerde Unsupervised yöntemlerden yararlanılabilir;** Eksik verinin olduğu kolon hesaba katılmadan diğer bağımsız değişkenlerle unsupervised kümeleme algoritmalarını kullanırsın. Eksik verinin bulunduğu kategorik verideki unique sınıf sayısı kadar küme oluşturması istenir algoritmadan. Daha sonra oluşturulan kümeler sınıflarla eşleştirilir ve eksik verilere denk gelen kümelere göre de kategorik sınıf belirlenip eksik verilere atanır.***(Eksik veri fazlaysa kullanıyorum)***

#### Median ile Doldurmak (Continous veriler)
* **Bu işlem için fillna() fonksiyonundan yararlanılır.**  


* Eksik değerleri değişkenin tüm değerlerinin Median'ı ile doldurmak yanlılık oluşturabileceği için veri setindeki herhangi bir kategorik veriye göre gruplama yapıp gruplara göre Medyan ortalaması ataması yapacağız.  


* ***!!! Train datasında uygulanan işlemin aynısı Validation ve Test setleri üzerinde de uygulanmak zorundadır. Train setinde kullanılan Median değeriyle Validation ve Test setindeki eksik veriler doldurulur. Train'de ne kullandıysan diğerlerine onu implement etmek ZORUNDASIN!!!***


Avantajları ve Dezavantajları;  

* Kolay uygulanabilir bir tekniktir.  

* **Median kullanımı outlier'lara karşı robust bir davranış gösterir.**  

* Verinin dağılımını ve korelasyonunu değiştirir.  

```python
df = pd.read_csv("train.csv")
df1 = df.copy()
```

```python
# Örneğimizde "Age" değişkenindeki eksik değerleri dolduracağız.
# Gruplama için "Pclass" kategorik değişkenini seçtik.
df.groupby("Pclass")["Age"].median()
```

```python
# Görüldüğü üzere tüm değerler doğru şekilde değiştirilmiş.
df.iloc[ix,:]
```

#### One Hot Encoding
***MANTIK;***  
* Nominal verilerde uygulanan bir yöntemdir. Nominal veriler; Erkek/kadın, Siyah/Beyaz/Kırmızı gibidir.  

  
* **Siyah/Beyaz/Kırmızı örneğini alacak olursak. Siyah'ın denk geldiği gözlemde Siyah için oluşturulan değişkende 1 değeri gösterilecek ve diğer renklerin aynı satırdaki gözlemleri 0 olacaktır. Kırmızıya denk gelen bir gözlemde de Kırmızının kolonunda 1 değeri görülürken diğerlerinde 0 görülecektir.**  


* Bu yöntemde değişkendeki sınıf sayısı kadar yeni değişken üretilir. Ancak bu yeni değişkenlerden bir tanesi silinmelidir, bunun nedeni son değişken olmasada verinin anlamını kaybetmeyeceği durumundan gelmektedir. Gereksiz değişkenlerin atılmasını istediğimiz için yeni oluşturulan değişkenlerden herhangi birini sileriz. Bu duruma Dummy Variable Problemi denilir.    


* **One Hot Encoding uyguladığın kolondaki değerler değişken nümerik olmamalıdır, hata verir.**  


* Üzerinde One Hot encoding işlemi yapılan kategorik değişkenin orijinal hali de silinebilir.


***ÖZELLİKLER;***  
* Her türlü ML probleminde kullanılabilir.  
* Nominal temelli bir tekniktir.  
* **Low Cardinality'li değişkenlerde kullanılması önerilir.** (*Kullanılması durumunda aşağıdaki 2 ve/veya 3 numaralı notu dikkate al*)  
* <span class="mark">Nominal ve low cardinality'li verilerde kullanılması tavsiye edilir.</span>  
* Tree based algoritmalarda iyi çalışmadığı için kullanılmaması öneriliyor.  (*Kullanılması durumunda 1 numaralı notu dikkate al*)  
* **Logistic Regresyon, Linear Regresyon gibi linear modellerde başarılı olduğu söyleniyor.**
* Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa bu gözlemlerin tüm One Hot Encoding kolonlarında değerleri 0 olarak atanmalıdır.  
* **Overfitting/Data Leakage eğilimi YOKTUR.**  


<span class="burk">***ÖZEL NOTLAR/PÜF NOKTALARI;***</span>  
1. <span class="mark">Eğer yine de ağaç tabanlı algoritmalarda kullanmayı planlıyorsan da tüm kolonları kullanman önerilir, yani bir tanesini silmeden model kurmayı denemen önerilir. Çünkü dummy variable aslında multicollinearity denilen problemi getiriyor ama tree based modellerde multicollinearity bir problem değil ancak regresyon problemlerinde ve diğer algoritma türlerinde dummy variable'ı yok etmen gerekecektir.</span>   
 

2. <span style='background :pink' >Kategorik değişkende çok fazla sayıda sınıf bulunduğunda(yani high cardinality varsa) One Hot Encoding işlemi çok fazla sayıda yeni değişken oluşturur, bu nedenle pek high cardinality durumunda pek tavsiye edilmez. Bu durum çok istenen bir durum değildir, çünkü curse of dimensionality problemine yol açmaktadır. Bunu önlemek için değişkende frekansı en fazla olan ilk 10-15 sınıf tutulur ve diğerleri sanki aynı sınıfı temsil ediyormuş gibi "others" olarak isimlendirilir. Bu düzenlemeyi yaptıktan sonra One Hot Encoding uygulanır. Böylece belki 50-60 tane yeni değişken oluşturulacakken 10-15 tane değişkenle bu işi tamamlamış oluruz.</span>

#### Target Mean Encoding
***MANTIK;***  

* Train setinde söz konusu kategorik değişkenin her bir sınıfına denk gelen Output değişkenindeki değerlerin ortalamasını alıyorsun ve kategorik değişkeninin sınıflarına atıyorsun.  


* **Train setinde bulduğun ortalama değerleri değiştirmeden Validation ve Test setinde de kullanıyorsun. Yani aynı şekilde ortalama değerleri Validation ve Test setinde de bulmak gibi bir işe sakın kalkışma.**  
  

***ÖZELLİKLER;***  
* Her türlü ML probleminde kullanılabilir. Yalnızca uygulanış biçimi Multiclass Classification problemlerinde biraz farklı olmalıdır. Aksi taktirde hedef değişkendeki sınıfların ortalamaları ML modelini şaşırtabilir. Öyle ki hedef değişkende 1 ve 3 değerlerini alan kategorik sınıfın ortalaması 2 olur ama aslında o kategorik sınıf hiçbir zaman 2 olmamıştır. İşte bunun önüne geçebilmek için önce hedef değişkene One Hot Encoding uygulanır(bu işlem sadece mean encoding için yapılır, daha sonra orijinal hedef değişken kullanılacaktır) daha sonra kategorik değişkenin sınıfları One Hot Encoding'le oluşturulan hedef değişkenin kolonlarına göre ayrı ayrı Mean Encoding işlemine tabi tutulur. Bu işlemin sonucunda Hedef değişkendeki sınıf sayısı kadar Mean Encoding kolonu oluşturulmuş olacaktır ve kafa karışıklığı ortadan kalkmış olacaktır.  


* Target(hedef değişken) temelli bir tekniktir. Literatürde Target Based encoder'lara Bayesian Encoder'lar isminin verildiği görülmüştür.   


* **Söz konusu kategorik değişkenle hedef değişken arasında güçlü bir ilişki varsa bu yöntemin kullanılması ya da en azından denenmesi tavsiye edilir.**  


* High cardinality'li kategorik değişkenler için çok sık kullanılan güzel bir tekniktir. Aynı şekilde Low Cardinality'li değişkenlerde de kulanılabilir ama daha az kullanılır.  


* **Her türlü algoritmayla kullanılabilir.**   


* **Binary Classification problemlerinde** Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa bu gözlemlerin Mean Encoding değeri 0.5 olarak atanmalıdır. Çünkü binary bir olayın gerçekleşme olasılığı %50'dir.  


* **Multiclass Classification problemlerinde** Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa Mean Encoding değeri; 1/n olmalıdır. (n= hedef değişkendeki farklı sınıf sayısıdır)  


* **Regresyon Problemlerinde** Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa Mean Encoding değerine bir şey yazamayız. Mantıksız olur. Bu nedenle eğer ki Validation ve Test setinde Train'de hiç olmayan bir değer bulnurusa bu yöntemi kullanma.  


* **Bu teknik Overfitting'e neden olabilir çünkü hedef değişkenle yüksek korelasyona sahip olacak bir değişken oluşturulmuş olunuyor. Dolayısıyla overfitting'i elimine edecek bir regularization algoritmasının kullanılması önerilir, random gürültü ekleyerek veya sıkı cross validation işlemleri uygulanması önerilir.**  


* Ayrıca Target Mean Encoding yerine Catboost Target Mean Encoding, K-fold Target Mean Encoding gibi overfitting'e robust olan diğer encoding yöntemleri kullanılarak overfitting/leakage probleminin önüne geçilebilir.

#### Weight Of Evidence Encoding (WOE)
***MANTIK;***  
* Kısaca ln(p(1)/p(0)). Yani kategorik değişkenin sınıflarının output değişkeninde 1 olma olasılığı ile 0 olma olasılığının oranının doğal logaritmasını alıyorsun.  


* Train setinde elde ettiğin değerleri kullanarak Validation ve Test setlerini de encode ediyorsun.


* MultiClass Classification problemlerinde Output'a One Hot Encoding uygulayıp WOE'leri her bir output sonucu için bulmaya çalış. Daha sonra One Hot Encoding'le oluşturulan kolonları silersin. Örnek olarak 4 sonuçlu bir multiclass classification probleminde 1 olma olasılığı ile 2,3,4(yani 1 olmama olasılığı: 1-p(1)) olma olasılığının oranlarının doğal logaritmasını alırsın. Daha sonra 2 olma olasılığı ile 1,3,4(yani 2 olmama olasılığı: 1-p(2)) olma olasılığının oranlarının doğal logaritmasını alırsın. Bu şekilde Output'ta bulunan her target sınıfı için WOE katsayılarını bulup encoding işlemini tamamlarsın.  


***ÖZELLİKLER;***  
* **Classification problemlerinde kullanılır.**  


* Targer(hedef değişken) temelli bir tekniktir. Literatürde Target Based encoder'lara Bayesian Encoder'lar isminin verildiği görülmüştür.   


* Söz konusu kategorik değişkenle hedef değişken arasında güçlü bir ilişki varsa bu yöntemin kullanılması ya da en azından denenmesi tavsiye edilir.


* High cardinality'li kategorik değişkenler için çok sık kullanılan güzel bir tekniktir. Aynı şekilde Low Cardinality'li değişkenlerde de kulanılabilir ama daha az kullanılır.


* Classificaton algoritmalarıyla kullanılabilir. **Logistic Regression** ile iyi sonuçlar verebilir.  


* **Binary Classification problemlerinde** Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa bu gözlemlerin WOE değeri 0 olarak atanmalıdır. Çünkü Binary bir olayın doğal olasılığı 0.5'tir. Hem olma hem de olmama olasılığının birbirine eşit olmasından dolayı olasılıkların oranı 1'dir ve ln(1)=0'dır. İşte bu nedenle Binary Classification problemlerinde yeni sınıfa "0" yazılır.  


* **Multiclass Classification problemlerinde** Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa WOE değeri; ln((1/n)/(1-1/n)) olmalıdır. (n= hedef değişkendeki farklı sınıf sayısıdır)  


*  Bu teknik Overfitting'e neden olabilir çünkü hedef değişkenle yüksek korelasyona sahip olacak bir değişken oluşturulmuş olunuyor. Dolayısıyla overfitting'i elimine edecek bir regularization algoritmasının kullanılması önerilir, random gürültü ekleyerek veya sıkı cross validation işlemleri uygulanması önerilir.  


* Ayrıca WOE Encoding yerine Catboost Target Mean Encoding, K-fold Target Mean Encoding overfitting'e robust olan diğer encoding yöntemleri kullanılarak overfitting/leakage probleminin önüne geçilebilir.

#### M-Estimate Encoding
***MANTIK;*** 
* Target Mean Encoding'in overfitting'e yatkın olma dezavantajını önlemek amacıyla benzer mantıkla uygulanan farklı bir işlemdir.  


* Bu yöntemde M isimli bir katsayı kullanılarak bir tür smoothing yapılır.  


* İşlemin formülü; [count(categor) * mean(category)] + [M * mean(target)] / [count(category) + M]  


* **Mutliclass Classification problemlerinde** target değişkenine One Hot Encoding(OHE) uygulanmalı ve hesaplamalar target değişkendeki OHE işlemi ile oluşturulan kolonlar için ayrı ayrı yapılmalıdır.  


* Train setinde bulduğun M-estimator değelerini Validation ve Test setlerinde de kullanıyorsun.  


***ÖZELLİKLER;***  
* Her türlü problemde kullanılabilir. ***Ancak bir Regresyon probleminde Train setinde olmayan bir sınıf Validation veya Test setinde bulunuyorsa bu durumda M-estimator encoding kullanmak çok mantıklı olmayacaktır. Çünkü "mean(category)" kısmına mantıklı bir şey yazamayız.***  


* Target(hedef değişken) temelli bir tekniktir. Literatürde Target Based encoder'lara Bayesian Encoder'lar isminin verildiği görülmüştür.     


* Söz konusu kategorik değişkenle hedef değişken arasında güçlü bir ilişki varsa bu yöntemin kullanılması ya da en azından denenmesi tavsiye edilir.


* High cardinality'li kategorik değişkenler için çok sık kullanılan güzel bir tekniktir. Aynı şekilde Low Cardinality'li değişkenlerde de kulanılabilir ama daha az kullanılır.  


* Her türlü algoritmayla kullanabilirsin.  


* **Binary Classification problemlerinde** Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa bu gözlemlerin M-estimator encoding değeri hesaplanırken mean(category) değeri 0.5 olarak alınmalıdır. Çünkü iki sonuçlu bir olayın olma olasılığı 0.5'tir, aynı paranın yazı-tura gelmesi gibi.  


* **Multiclass Classification problemlerinde** Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa M-estimator encoding değeri hesaplanırken mean(category) "1/n" olarak alınmalıdır. (n= hedef değişkendeki farklı sınıf sayısıdır)  


* <span class="mark">Her ne kadar overfitting'i engellemeye çalışsa da yine de overfitting riski göz önünde bulundurulmalıdır.</span>

2. **m =** M katsayısı. M katsayısı arttıkça smoothing artar, değişkenin tahmine etkisi azalır, ancak overfitting'den uzaklaşmış olur. M, negatif olamaz. M'in 1-100 arasında olması öneriliyor.  **Default=1.0, istersen bu şekilde bırakabilirsin veya farklı M katsayılarıyla denemeler yapıp en başarılı olanı kullanabilirsin.**

#### K-Fold Target Mean Encoding
***MANTIK;***  
* Mantık Target Mean Encoding'e benzer. Ancak bu sefer Target Mean değerleri hesaplanmadan önce veriseti fold'lara bölünür ve her fold için kendi içinde Target Mean değerleri hesaplanır.  


* Bu fold'lara bölünerek yapılan hesaplama işleri Target Mean Encoding yönteminin overfitting/leakage problemini de olabildiğince azaltmaktadır ve daha güvenilir bir sonuç ortaya koymaktadır.  


* Fold'lara bölme işlemi sadece Train seti üzerinde uygulanır. Validation ve Test setini bölmüyoruz.  


* Foldlara böldüğümüz verisetinde önce ilk fold'u devre dışı bırakıp diğer kalan fold'larla encode etmek istediğimiz kategorik değişkendeki sınıflara göre target mean değerlerini hesaplıyoruz ve sonrasında bulduğumuz mean değerlerini en başta devre dışı bıraktığımız fold'daki kategorik sınıflara uygun şekilde atıyoruz. Daha sonra 2. fold'a geçiyoruz ve 2. fold'u da devre dışı bırakıp diğer fold'larla kategorik sınıflara göre target mean değerlerini bulup bu değerleri hesaplamalarda devre dışı bıraktığımız fold'daki kategorik sınıflara atıyoruz. Aynı işlemleri her bir fold üzerinde yapıyoruz.  


* **Yukarıda bahsi geçen hesaplamalar Train seti üzerinde yapılır**. Test ve Validation seti için ise, Train setindeki fold'larla her bir kategorik sınıf için hesaplanan Target Mean değerlerinin ortalaması alınır ve bu ortalamalar direkt Test ve Validation setinde söz konusu kategorik değişkendeki sınıflara atanır.  


***ÖZELLİKLER;***  
* Her tür problemde uygulanabilir. Multiclass Classification problemlerinde kfold encoding işlemini yaparken Target değişkeninin One Hot Encoding ile encode edilerek her bir target sınıfına göre işlemlerin yapılması gerekmektedir.  


* Target based bir tekniktir.


* Her tür algoritmayla kullanılabilir.


* Binary Classification problemlerinde Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa bu gözlemlerin Mean Encoding değeri 0.5 olarak atanmalıdır. Çünkü binary bir olayın gerçekleşme olasılığı %50'dir.



* Multiclass Classification problemlerinde Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa Mean Encoding değeri; 1/n olmalıdır. (n= hedef değişkendeki farklı sınıf sayısıdır)



* Regresyon Problemlerinde Train setinde hiç bulunmayan bir sınıf Validation veya Test setlerinde bulunursa Mean Encoding değerine bir şey yazamayız. Mantıksız olur. Bu nedenle eğer ki Validation ve Test setinde Train'de hiç olmayan bir değer bulnurusa bu yöntemi kullanma.


* <span class="mark">Bu yöntemde Overfitting/Leakage problemleri olabildiğince önlenmeye çalışılmıştır. Bu nedenle güvenilir bir yöntem diyebiliriz.</span>

* Bu yöntem için Python'da özel bir fonksiyon yoktur, kendi kodumuzu yazmak gerekir.

* KFold fonksiyonunun önemli parametreleri;  
 **n_splits=** Veriyi kaç parçaya/fold böleceğini belirtir.  
 **shuffle=** Veriyi bölmeden önce shuffle eder, aslında biz Train/Validation/Test ayrımında verimizi shuffle ediyoruz ama rastgeleliği sağlamak adına bir kez daha burada shuffle edilebilir. Etmesen de olabilir, tamamen sana kalmış. 

**KFold ile böldüğün verinin parçalarını aşağıdaki kodda gördüğün gibi fd1, fd2, fd3.... diye adlandırabilirsin.
Eğer; fd1[1] ifadesini kullanırsan böldüğün ilk parçada bulunan gözlemlerin index numaralarına ulaşırsın. fd1[0] ifadesini kullanırsan da böldüğün parça dışında diğer tüm veride kalan gözlemlerin index numaralarına ulaşırsın. Aynı şekilde fd2[1] yazarsan da 2.parçanın gözlemlerinin index numaralarına ulaşırsın. Bu index numaraları işlemleri yapmanı kolaylaştıracaktır.**

```python
# Önce verinin import edilmesi ve split edilmesi tamamlanır.
df = pd.read_csv("train.csv")
df1 = df.copy()



X = df.drop(["Survived"],axis=1)
y = df.Survived

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True, random_state=2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle=True, random_state=2)

for i in [X_train, X_val, X_test, y_train, y_val, y_test]:
    i.reset_index(inplace=True, drop=True)
```

```python
from sklearn.model_selection import KFold
fd1, fd2, fd3, fd4, fd5 = KFold(n_splits=5, shuffle=True, random_state=5).split(X_train)
```

```python
# Önce kfold encoding için Train setinde boş kolonlar oluşturuyorum.
for i in ["Sex","Embarked","Pclass"]:
    X_train[i+"_kfold_mean"] = X_train[i]


    
# Train setinde k-fold'u uyguluyorum;    
for i in [fd1, fd2, fd3, fd4, fd5]:
    for j in ["Sex","Embarked","Pclass"]:
        kfold_mean = pd.concat([X_train.iloc[i[0]],y_train.iloc[i[0]]],
                               axis=1).groupby(j)["Survived"].mean().to_dict()
        X_train[j+"_kfold_mean"].iloc[i[1]] = X_train[j+"_kfold_mean"].iloc[i[1]].map(kfold_mean)



# Yukarıdaki for döngüsündeki son "map" işleminden sonra söz konusu değişkenlerin veri tipi "object" olarak kaydediliyor.
# Ancak aşağıdaki işleme devam edebilmem için bu değişkenlerin "float" olması lazım. Bu yüzden aşağıdaki işlem yapılmıştır.
for i in ["Sex","Embarked","Pclass"]:
    X_train[i+"_kfold_mean"] = X_train[i+"_kfold_mean"].apply(float)



# Train setinde bulduğum k-fold değerlerinin ortalamalarını Validation ve Test setinde kategorik değişkendeki sınıflara atıyorum
for i in [X_val, X_test]:
    for j in ["Sex","Embarked","Pclass"]:
        kfold_mean = X_train.groupby(j)[j+"_kfold_mean"].mean().to_dict()
        i[j+"_kfold_mean"] = i[j].map(kfold_mean)
```

#### Binary Encoding

* Temel olarak kategorik değişkendeki her bir sınıfın binary kodlara dönüştürülmesine dayanan bir encoding yöntemidir.  


* Feature Hashing yöntemine çok benzer, ancak Feature Hashing yönteminde encoding işlemi sonucu yeni oluşturulucak encoding kolonlarının sayısı kullanıcı tarafından belirlenebilirken Binary Encoding yönteminde bunu kullanıcı belirleyemez.  


* Binary Encoding fonksiyonu üzerinde çalıştığı kategorik değişkendeki unique sınıf sayısının 2 tabanındaki logaritması kadar yeni kolon oluşturur. Örneğin 100 adet farklı kategorik sınıfı içeren bir değişken için Binary Encoding işlemi sonucunda 7 tane yeni kolon oluşturulmuş olur. #### Helmert Encoding
***MANTIK;***  
* *Bu yöntemi anlamam biraz zaman aldı zamanında ama gerçekten kolay bir yöntemdir.*  


* Mantık Ordinal Encoding ile One Hot Encoding(OHE) kombinasyonuna benzer.  


* Bu yöntem sınıfların arasında ordinal bir ilişkinin olduğu kategorik değişkenlerde başarılıdır.  


* İlk olarak encode edeceğimiz kategorik değişkeni küçükten büyüğe olacak şekilde sort ederiz. Bunları nasıl yaptığım konunun sonundaki kodlarla çok güzel bir şekilde örneklendirilmiştir.  


* Daha sonra Helmert Encoding fonksiyonunu çalıştırırız ve sonucumuzu elde ederiz.  


* Helmert Encoder aynı OHE gibi kategorik değişkendeki sınıf sayısının 1 eksiği kadar yeni kolon oluşturacaktır.  


* Peki bu sonuç ne anlama geliyor???  

1. En başta ne dedik? Ordinal kategorik değişkenimizi ordinal sıraya göre küçükten büyüğe olacak şekilde sort edelim.

2. Helmert Encoder, en başta olan yani ordinal olarak en küçük olan kategorik sınıfı oluşturduğu tüm kolonlarda "-1" olarak gösteriyor ve bu sınıfa ait bir kolon da oluşturmuyor. Yani dummy variable mantığı gereği veriden atılan kolon bu sınıfa ait olan kolon oluyor. Helmert Encoder bundan sonra karşılaşacağı sınıflar için sırayla kolonlar oluşturuyor.  

3. Karşılaştığı bir sonraki sınıf için bu sınıfın kendi kolonunda "1" gösteriyor, ancak bu sınıftan ordinal olarak daha büyük olan sınıfların kolonlarında bu değere karşılık "-1" gösteriyor. Örneğin Titanic setindeki 2.sınıf bileti olan yolcular Helmert Encoder'ın 1.sınıflar için oluşturduğu kolonda "-1" olarak gözükür.  

4. Gelelim bir sonraki sınıfa. Helmert Encoder bu sınıfı kendi kolonunda "2" olarak gösteriyor ve bu sınıfı kendisinden ordinal olarak büyük olan sınıfların kolonlarında "-1" olarak gösteriyor, bununla birlikte kendisinden ordinal olarak küçük olan sınıfın kolonunda da "0" olarak gösteriyor.  

5. Geçelim bir sonraki kategorik sınıfa. Helmert Encoder bu sınıfı kendi kolonunda "3" olarak gösteriyor ve bu sınıfı kendisinden ordinal olarak büyük olan sınıfların kolonlarında "-1", kendisinden ordinal olarak küçük olan sınıfların kolonlarında da "0" olarak gösteriyor.  

6. İşte Helmert Encoder mantığı bu şekilde ilerliyor.  


7. Bizim kategorik değişkeni en başta ordinal olarak küçükten büyüğe sort etmemizin sebebi de bu ordinal sıralamanın bozulmadan korunmasındandır. Aksi taktirde sıralama karışık olabilir ki bu da istediğimiz şeyi yakalamamıza engel olur.  


***ÖZELLİKLER;***  
* Her tür problemde kullanılabilir ancak özellikle Regresyonda yaygın olarak kullanılır. Regresyonda OHE ve Effect Coding'den sonra en yaygın kullanılan encoding yöntemidir.

#### Cycle Encoding
***MANTIK;***  
* Bu teknik döngüsel verilerde kullanılır. Yani ay, hafta, gün, saat, dakika, saniye verileri gibi olan değişkenlerde.  


* Örneğin saatleri ele aldığımız zaman saat 22 ile 23'ün farkı 1 saat ama 23 ile gece yarısını 23-0 olarak hesaplayıp farkı 23 saat gibi algılayabiliyor ML algoritmaları. İşte bu durumu düzeltmek için Cycle Encoding güzel bir yöntem olarak karşımıza çıkıyor.  


* Bu iş için döngüsel bir fonksiyon olan cosine fonksiyonundan yararlanıyoruz. Yalnız cosine fonksiyonu tek başına yeterli olmuyor. Çünkü; cosine fonksiyonunda bazı sınıflar eşit cosine değerleri alıyor ve bu durum ML algoritmasının bu sınıfları sanki aynıymış gibi algılamasına neden oluyor. İşte bu durumu önlemek için encoding işlemimize ikinci bir dimension daha ekliyoruz. Bu dimension da döngüsel bir fonksiyon olan sine fonksiyonudur.  

* cosine için; cos([2*pi*X] / max(X))  
* sine için; sin([2*pi*X] / max(X)) formülleri uygulanır.  


* Ay, hafta, gün gibi verilerde değerleri 0'dan başlatman öneriliyor. Saat, dakika, saniye verilerinde veri zaten 0'dan başlıyor. Bu 0'dan başlatmanın nedeni döngünün doğru şekilde hesaplanmasını sağlamak içindir. Örneğin; Ocak=0, Şubat=1, MArt=2 olacak şekilde bunlar numaralandırıp daha sonra sine ve cosine dönüşümlerini yapman gerekiyor.  


* Pyhton'da bu yöntemi uygulayacak özel bir fonksiyon yok. Manuel olarak bir işlem kodu yazacağız. Aynı kodu Train, Validation ve Test setlerinde uygulayacaksın. Gayet kolay bir yöntem.

* **Decision Tree based algoritmalarda tam verimli çalışmama ihtimali olduğu söyleniyor. Çünkü ağaçlardaki leaf'lar kriterlerini multivariate kuramadıklarını dolayısıyla sine ve cosine değerlerini ortak şekilde irdeleyemeyeceği belirtiliyor. Aslında bana göre decision tree algoritmaları bunu yapabilir ama bu konuda tam emin değilim. Bununla birlikte böyle bir durumda cosine ve sine değerlerinin PCA/LDA gibi bir yöntemle tek bir kolonda birleştirilerek decision tree algoritmalarına verilmesi de bir çözüm olabilir. Denemedim bilmiyorum.**

2. Exponantial
3. Kök alma/ dereceli kök alma
4. Logaritma alma (log1p isimli numpy fonksiyonunu kullanacağız. Böylece logaritma 0 olduğunda çalışacak çünkü bu fonksiyon veriye "1" ekliyor daha sonra logaritmasını alıyor)
5. Reciprocal/ dereceli reciprocal 
6. Box-Cox Dönüşümü

```python
# İşlemlerin uygulanabilmesi için değişkenimizde eksik veri olmamalı o nedenle rastgele olarak 20 ile eksik verileri doldurduk
X_train["Age1"] = X_train.Age.fillna(20)
X_val["Age1"] = X_val.Age.fillna(20)
X_test["Age1"] = X_test.Age.fillna(20)


# Aşağıdaki örnekte verinin Left veya Right skewed olmasına bakılmaksızın tüm örnekler aynı değişken üzerinde denenmiştir.
# Bu durum tamamen örnek amaçlıdır. Normalde skewness yönüne bakmak önemlidir.  

# 1.Üs Alma;
X_train["Age_us_alma"] = X_train.Age1.apply(lambda x: x**5)
X_val["Age_us_alma"] = X_val.Age1.apply(lambda x: x**5)
X_test["Age_us_alma"] = X_test.Age1.apply(lambda x: x**5)

# 2.Exponantial;
X_train["Age_exp"] = X_train.Age1.apply(lambda x: np.exp(x))
X_val["Age_exp"] = X_val.Age1.apply(lambda x: np.exp(x))
X_test["Age_exp"] = X_test.Age1.apply(lambda x: np.exp(x))

# 3.Kök alma(5.dereceden kök alınmıştır);
X_train["Age_root"] = X_train.Age1.apply(lambda x: x**(-1/5))
X_val["Age_root"] = X_val.Age1.apply(lambda x: x**(-1/5))
X_test["Age_root"] = X_test.Age1.apply(lambda x: x**(-1/5))

# Logaritma alma;
X_train["Age_log"] = X_train.Age1.apply(lambda x: np.log1p(x))
X_val["Age_log"] = X_val.Age1.apply(lambda x: np.log1p(x))
X_test["Age_log"] = X_test.Age1.apply(lambda x: np.log1p(x))

# Reciprocal(8.dereceden aldık örnek olarak);
X_train["Age_reciprocal"] = X_train.Age1.apply(lambda x: 1/(x**8))
X_val["Age_reciprocal"] = X_val.Age1.apply(lambda x: 1/(x**8))
X_test["Age_reciprocal"] = X_test.Age1.apply(lambda x: 1/(x**8))

# Box-Cox Yöntemi;
# Box-Cox yöntemi önce Train setine uygulanır. Fonksiyon Train setini transform ederken en uygun lambda değerini de bulur.
# Daha sonra bu bulunan lambda değeri değiştirilmeden Box-Cox transformasyonu Validation ve Test setlerinde de uygulanır.
from scipy import stats
X_train,fitted_lambda = stats.boxcox(X_train.Age1)
X_val = stats.boxcox(X_val.Age1, fitted_lambda)
X_test = stats.boxcox(X_test.Age1, fitted_lambda)
```

* Outlier'ların varlığı bazı ML algoritmalarının performansını olumsuz etkilerken bazılarına herhangi bir etki yapmamaktadır.  


* Özellikle Fraud projelerinde Outlier'ların varlığı işe yaramaktadır.

* Outlier'lara müdahale ederken uyguladığım silme/baskılama gibi teknikleri sadece **Train** setinde uyguluyorum. Diğerler Validation ve Test setlerinde baskılama veya silme işlemleri yapmıyorum. **Ancak yeni değişken oluşturma gibi bir teknik uygularsam bunu hem Validation hem de Test setlerinde de uygulamak zorundayım.**

#### IQR Yöntemi ile (Normal Dağılımlı olmayan verilerde kullanılır);
* IQR = Quantile(0.75) - Quantile(0.25)  


* IQR değeri yukarıdaki eşitlikle bulunur. Daha sonra IQR ile 1.5 çarpılarak yeni bir değer bulunur.  Bu değeri de Quantile(0.25) değerinden çıkarıp, Quantile(0.75) değerine de ekleyerek iki değer daha elde edilir ve bunlar bizim outlier'lar için belirlediğimiz üst ve alt sınırlar olur.  


* Daha sonra bu sınırların dışında kalan gözlemler outlier olarak kabul edilir.  


* Kimi yerlerde IQR değeri 1.5 yerine 3 ile çarpılarak sınırlar bulunur. Bu şekilde belirlenen sınırların dışında kalan gözlemlere "extreme outliers" denilir.  


***Not; Bu yöntemde outlier olarak tespit edilmiş değerleri scatter plot/histogram gibi grafiklerde kontrol edin. Aslında multivariate gözlemlerde outlier olarak görünmeyen verileri outlier olarak tespit etmiş olabilirsiniz. Bu durum veri kaybı yaşamanıza ya da veriyi yanlış anlamanıza sebep olur. Her zaman görsel grafiklerle tespit ettiğiniz outlierların uzaysal boyuttaki pozisyonlarını inceleyin.***

**Yapılabilecek Müdahaleler;**  
* Tespit ettiğin outlier'ları veriden silebilirsin.**Çok tercih etmiyorum**       


* Outlier'ları yeni bir değişken oluşturarak etiketleyebilirsin. Bunu yaparsan Validation ve Test seti için de yeni değişken oluşturman gerekir. Validation ve Test setinde de bu yeni değişkeni oluştururken Train setini analiz ederken bulduğun IQR sınırlarını kullanmalısın.    


* Outlier'ları IQR sınırlarının değerlerine eşitleyebilirsin.**Ben bunu sevmiyorum, çünkü bu teknik yanlılık oluşturabiliyor ama proje gereği illa ki outlier'ların etkisinden kurtulmam gerekiyorsa bu yöntem tercih edilebilir.**

```python
# Bu sefer de "carat" isimli değişkenle çalışalım
sns.distplot(X_train["carat"], fit=norm);
```

```python
upper_iqr_limit = X_train.carat.quantile(0.75) + ((X_train.carat.quantile(0.75)-X_train.carat.quantile(0.25))*1.5)
lower_iqr_limit = X_train.carat.quantile(0.25) - ((X_train.carat.quantile(0.75)-X_train.carat.quantile(0.25))*1.5)

carat_outliers = X_train["carat"][(X_train.carat >= upper_iqr_limit) | (X_train.carat <= lower_iqr_limit)]
carat_outliers # "carat" değişkenine ait outlierlar bu şekilde bulunmuş olur.
```

```python
# kontrol amaçlı bir değişken oluşturalım, bu değişken "carat" değişkeninde outlier olan gözlemleri göstersin.
X_train["carat_outlier"] = "Not Outlier"
X_train["carat_outlier"][carat_outliers.index] = "Outlier"
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10,10))
sns.histplot(data=X_train, x="carat", hue="carat_outlier", ax=axes[0])
sns.scatterplot(data=X_train, x="table", y="carat", hue="carat_outlier", ax=axes[1]);
# Grafiklere bakıldığında nispeten doğru bulmuşuz ama belki burada extreme outlier'ları almak daha doğru olabilir.
# Çünkü özellikle scatterplot'a bakıldığında işe yarayacak verilerin de Outlier sınırına girdiği görülmekte.
# Extreme outlierlar IQR*3 ile bulunan sınırların dışında kalan gözlemlere denilmektedir.
```

* Bir verisetindeki bütün değişkenleri dahil ederek outlier'ları bulmak isterken değişkenlerin nümerik veya kategorik olması farketmez, her türlü veri tipiyle çalışılabilir. **Yalnız, kategorikleri encode etmen gereklidir!**


* Verinizi Scale etmenize de gerek yoktur.  


* Isolation Forest ağaç tabanlı bir algoritma olup Fraud detection problemlerinde de bir classification algoritması olarak kullanılır. Başarılı sonuçlar da verir.  


* Bu yöntemle outlier'ları tespit ederken, Train seti kullanılarak bir Isolation Forest modeli oluşturulur. Bu model Train setindeki outlier'ların hangi gözlemlerde olduğunu tahmin eder.  


* ***Algoritmanın "1" olarak tanımladığı değerler inlier/normal gözlemlerdir. "-1" olarak tanımlanan değerler ise Outlier'ları temsil etmektedir.***  

**Yapılabilecek Müdahaleler;**  

* Tespit ettiğin outlier'ları veriden silebilirsin.**Çok tercih etmiyorum**  


* **Tüyo-1:** Algoritmanın bulduğu outlier'lar sanki bir değişkenmiş gibi verisetine bir feature olarak eklenebilir. Yalnız böyle bir durumda benzer değişkenleri Validation ve Test setleri için de oluşturmalısın. Burada IsolationForest algoritmasını **Train seti ile fit etmelisin** ve sonra oluşturduğun modeli **Validation ve Test setinde predict** yapacak şekilde kullanmalısın.   


* **Tüyo-2:** Hatta algoritma birgözlemi outlier olarak değerlendirmeden önce o gözleme bir skor verir, bu skor -0.5'ten küçükse(bu sınır değiştirilebilir, yani -0.5 yerine -0.4 de kullanılabilir) o gözlem outlier olarak nitelendirilir. İşte bu noktada belki de algoritmanın gözlemlere verdiği skorlar kullanılarak yeni bir değişken oluşturulabilir. Eminim ki bu imbalanced classification problemlerinde çok yardımcı olacaktır. Çünkü böyle problemlerde hedef değişkendeki minor sınıflar aslında birer outlier/fraud olarak davranmaktadır ve bunların IsolationForest score'ları muhtemelen -1'e yakın olacaktır.

**Isolation Forest Teori;**  
1. Bu algoritma ağaç tabanlı olup Random Forest'a benzer şekilde çalışmaktadır.  


2. Isolation Forest'ta bir sürü ağaç oluşturuluyor. Bu ağaçların her birinde verisetinden random şekilde örneklemler seçiliyor ve yine random şekilde verisetinden bir feature veya featurelar seçiliyor.


3. Daha sonra bu ağaçlar split(yani ağaç dallandırılıyor) edilerek ağaç yapısı oluşturuluyor. ***Split etme işlemi örneklemdeki her bir gözlem(observation) kendi bölgesinde tek başına kalana kadar(yani isolated olana kadar) sürdürülüyor.***  


4. Daha sonra tüm verisetindeki gözlemler incelenmeye başlanıyor. Gözlemlerin oluşturulan ağaçlarda ortalama kaçıncı splitten sonra isolated oldukları hesaplanıyor.  


5. Isolated olması için ortalama split sayısı az olanlar(örneğin 4 splitte isolated edilmiş) outlier olarak tanımlanıyor. Çünkü outlier'lar zaten veri kümesinin dışında olduğu ve diğer gözlemlerden uzakta olduğu için ayrıştırılması daha kolay olan gözlemlerdir.  


6. Benzer şekilde isolated olması için ortalama split sayısı daha fazla olanlar(örneğin 20 split) da inlier/normal gözlem olarak nitelendiriliyor. Çünkü veri kümesinin olasılık olarak kalabalık kısmını temsil eden bu değerleri ayrıştırmak için epey bir split yapmak gerekir.  


7. İşte bu şekilde her bir gözlemin isolated olması için gereken ortalama split sayısı hesaplanıyor ve bu gözlemlere bir skor veriliyor. Bu skor -0.5'ten küçükse o gözlem outlier olarak nitelendiriliyor. Bu -0.5 eşik değeri istenildiğinde değiştirilebilir.  


**Algoritmanın Önemli Hiperparametreleri;**  

* **Contamination:** Verideki gözlem sayısı göz önünde bulundurularak, gözlemlerin yüzde kaçının Outlier olarak tanımlanacağını belirtir. **Default olarak 0.1 verilmiş ama outlier detection amacıyla bu işi yaparken belki daha düşük bir contamination kullanılabilir. Çünkü bir verisetinin %10'unun outlier olması pek de sağlıklı olmayabilir.** Ayrıca bu parametre kafa karıştırmasın; şöyle ki contamination'a verilen değer doğrultusunda outlier buluyoruz ama eğer contamination'ı çok vermişsek esasında o kadar da outlier olmayan gözlemlerin de outlier olarak nitelendirildiğini görürüz. Bu durumu daha iyi açıklamak adına algoritmanın gözlemlere atadığı skorları incelemekte fayda var. Outlier olarak nitelendirilen ama aslında pek de outlier olmayan gözlemlerin algoritmanın tarafından atanan skorlarının -0.5'e yakın değerler aldığını göreceğiz. Böylece bu değerleri istersek daha sonra bu değerlere outlier değillermiş gibi muamele yapabiliriz. Isolation Forest algoritması gözlemlere skor verirken contamination değerine bakar ve ayarlamalarını öyle yapar. **Kısacası gözlemlerin score'larını incelemek bence çok gerekli.**  


* **n_estimators:** Kaç tane ağaç kurulacağını belirten parametredir. Default olarak 100 verilmiş, isteğe bağlı olarak arttırılabilir. **Ben default değeri kullanıyorum. Algoritmanın bariz outlier'ları gözden kaçırdığı durumlarda bu parametre arttırılabilir, böylece algoritmanın gözden kaçan outlier'ları yakalama şansı artar.** 


* **max_feature:** Ağaçlarda kaç tane feature kullanılacağını belirten parametredir. Default=1, yani univariate bir analiz söz konusudur. Eğer 2 veya 3 yapılırsa multivariate bir analiz yapılmış olabilir. **Ben 3'ü tercih ediyorum, bu muhtemelen analiz süresini uzatacaktır.**


* **max_samples:** Ağaçların kurulduğu örneklemlerin kaç adet gözlem içereceğini belirtir. Kullanılan değer verisetindeki toplam gözlem sayısından büyük olursa ağaçlarda tüm gözlemler kullanılır. **Default="auto", ama ben bu parametreye verisetindeki gözlem sayısını veya daha büyüğünü yazıyorum ki algoritma tüm gözlemleri bir arada değerlendirsin. Ve gerçekten de bu uygulamam çok işe yarıyor, çünkü öbür türlü algoritma bazı bariz outlier'ları gözden kaçırabiliyor.**


* **bootstrap:** Bu özellik True olursa verisetinden farklı farklı örneklemler alınır ve ağaçlar bunlar üzerinde oluşturulur. Bu özellik kapalı olursa ağaçlar aynı örneklem üzerinde çalıştırılır. **Ben False olarak kullanıyorum çünkü benim örneklemlerim  max_samples parametresinde yaptığım ayarlamaya göre tüm verisetini kapsıyor.**


* **random_state:** oluşturulan modele verdiğin ID numarası gibi bir şey.

```python
# Önce verimizi çağıralım;
df = pd.read_csv("diamonds.csv")
df1 = df.copy()

# Veriseti X=bağımsız değişkenler ve Y=output olarak ayrılır.
# Isolation Forest algoritması kategorik değişkenlerin encode edilmesini ister, zaman kaybetmemek için ben kategorikleri sildim
X = df.drop(["price","clarity","color","index","cut"],axis=1)
y = df.price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True, stratify=y, random_state=2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle=True, stratify=y_train,
                                                  random_state=2)

# Shuffle özelliğini açtığımız için index'ler karışmış halde olacaktır. Indexleri düzene sokmak için şu kodları yaz;
for i in [X_train, X_val, X_test, y_train, y_val, y_test]:
    i.reset_index(inplace=True, drop=True)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(random_state=42, contamination=0.01, max_features= 3, bootstrap=False, max_samples=50000 )
outliers = iso.fit_predict(X_train)
outliers
```

```python
print("X_train'deki gözlem sayısı: ",X_train.shape)
print("Bulunan outlier sayısı: ", outliers[outliers==-1].size)
# Görüldüğü üzere satır sayısı yani gözlem sayısının %1'i kadar outlier bulundu.
# Outlier'ları grafiklerle de kontrol ederek contamination parametresini değiştirebilirsin.
```

```python
# Algoritmanın gözlemlere verdiği değerler, -0.5'den küçük olanlar outlier olarak tanımlanır.
outlier_scores = iso.score_samples(X_train)
outlier_scores
```

```python
# Gözlemlerin outlier olup olmadığını ve skorlarını gösteren değişkenleri oluşturabiliriz.
X_train = pd.concat([X_train, pd.DataFrame({"IsOutlier":outliers, "outlier_scores":outlier_scores})], axis=1)
X_train
```

## Feature Scaling

**1.Feature Scaling Nedir?**

* Değişkenlerin benzer ölçeklerde olmasını sağlamak için uygulanan yöntemlere feature scaling denilir.  


* <span class="mark">Feature Scaling işlemlerinin uygulanması her ML modeli için mecburi değildir. Uygulanmasını gerektiren ve gerektirmeyen durumlar vardır.</span>  


* Bu işlemlerde değişkenlerin ortalamaları, standard sapmaları değişebilir ama anlamları değişmez.  


* <span class="mark">Feature Scaling değişkenlerin olasılık dağılımlarını değiştirmez.</span>  


* <span style='background :yellow' >Scaling işlemleri sadece ""continous"" datalar üzerinde uygulanır. Discrete yapıdaki verileri bin'lere bölmeni tavsiye ederim. Öte yandan nominal veya ordinal kategorik verilerin hiçbir şekilde feature scaling işlemlerine tutulmasına gerek yoktur. Eğer Normalization işlemi yapıyorsan işleme 0-1 şeklindeki nominal kategorik verileri ekleyebilirsin çünkü bunlar normalization işleminden etkilenmezler, normalization sonrası değerler yine 0-1 olarak kalır. Ancak Standardization işlemine hiçbir türlü kategorik değişkenleri sokmamak gerekir.</span>


* <span class="mark">Scaling işlemleri bağımsız değişkenlere(input) uygulanır. Hedef değişkene(output) uygulanmaması önerilir.</span> Hedef değişkenin Normal Dağılımlı olması için çeşitli transformasyonlar uygulamanın bir mahsuru yoktur ama scaling uygulanması pek önerilmiyor.   


* Feature Scaling, feature engineering aşamalarının son adımı olarak uygulanmalıdır.  


* <span class="mark">Feature Scaling uygulanacağı zaman önce Train setine uygulanır. Daha sonra Train setine uygulandığı şekilde Validation ve Test setlerine de uygulanmak zorundadır.</span>  

**2.Neden Feature Scaling Uygulanır?**  

1. Elimizdeki verisetinde bulunan continous değişkenlerin değer aralıkları birbilerinden çok farklı olabilir. Örneğin; bir şirkette çalışan personelin yaşlarını gösteren değişken ile maaş değişkeninin değer aralıkları birbirinden farklıdır. Bu durum bazen ML algoritmalarında yanlılık denilen hatalı tahmin performansının ortaya çıkmasına sebep olabilir.  


2. Bir değişken 3000metreyi gösterirken diğer değişken 5km'yi gösteriyor olabilir. Bazen kullanılan ML algoritması aslında daha küçük bir değer olmasına rağmen 3000metreyi daha büyükmüş gibi algılar. İşte bu yanlış ve yanlı tahminlerin yapılmasına sebep olabilir. <span class="mark">Özetle; daha büyük ölçekli değişkenler küçük ölçekli değişkenleri baskılar.</span> 


3. En genel anlamda bazı Algoritmalar feature scaling işlemlerine ihtiyaç duyarlar, scaling sonucunda daha hızlı, daha efektif ve daha doğru sonuçlar verirler.  


4. Gradient Descent Temelli algoritmalarda değişkenlerde feature scaling uygulanırsa algoritma daha hızlı çalışır. Çünkü değişkenler benzer değer aralıklarında olduğu için step size daha küçük olur ve işlemin hızı artar. <span class="mark">Gradient Descent Temelli Algoritmalar; Linear Regression, Logistic Regression, Neural Network.</span>  


5. <span class="mark">Neaural Network için Normalization MUST olarak bilinir.</span>  


6. Distance Based Algoritmalar feature scaling işlemine ihtiyaç duyarlar. Çünkü bu algoritmalar gözlemlerin aralarındaki uzaklıklardan yararlanarak tahminlerini yapar. Daha büyük değerleri olan değişkenler sanki daha uzaktaymış gibi davranacaktır ve küçük değerlerden daha farklı bir şekilde sonuca etki edecektir. Eğer Feature Scaling uygulanırsa değişkenler eşit oranda sonuca etki edecektir. <span class="mark">Distance Based Algoritmalar; KNN, K-means, Supported Vector Machine.</span>  


7. <span class="mark">PCA algoritması da feature scaling işlemlerine ihtiyaç duymaktadır.</span>  


8. <span class="mark">Decision Tree, Boosting Algoritmaları (GBM, XGBOOST....), Random Forest gibi kural tabanlı algoritmalarda feature scaling uygulanmasına gerek yoktur. Çünkü herhangi bir değişiklik sağlamazlar.</span>  


9. <span class="mark">LDA ve Naive Bayes gibi algoritmalarda da feature scaling işlemlerinin bir işe yaramayacağı düşünülür. Bu nedenle bu algoritmalarda feature scaling uygulanmasına gerek yoktur.</span>  


**3.Feature Scaling Teknikleri Nelerdir?**  
1. Normalization  
2. Standardization  
3. Robust Scaler  


**NORMALIZATION;**  
* MinMax scaler olarak da bilinir.  
* Değişkenin değerlerini [0,1] aralığına ölçeklendirir. Yani değerler belirli bir aralığa scale edilir.  
* Normal dağılımlı olmayan değişkenlerde kullanılması önerilir.  
* Neural Network ve distance based (KNN...) algoritmalarda kullanılması önerilir.  
* Bu teknik outlier'lara karşı oldukça hassastır, etkilenir. Çünkü değerleri belli bir aralığa sığdırmaya çalışıyoruz ve outlier'lar diğer değerleri baskılayabilir.  
* Normalization değişkenin dağılımında şekil olarak bir değişiklik oluşturmaz.  
* 0-1 şeklindeki nominal kategorik verileri Normalization işlemine sokabilirsin. Aslında kategorik verilerde scaling işlemi kullanılmaz ama Normalization nominal kategorik verilerde hiçbir değişiklik yapamayacağı için onları da işleme dahil edebilirsin. Bir sıkıntı olmaz.  


**STANDARDIZATION;**  
* Z-Score Normalization olarak da bilinir.  
* Değişkenleri Standard Normal Dağılıma dönüştürür.  
* Yani; değerleri 0 noktasında merkezleyen, ortalamalarını 0 ve standart sapmalarını 1 yapan bir scaling tekniğidir.  
* Bu tekniğin Normal Dağılımlı değişkenlerle kullanılması önerilir. Çünkü Standard Normal Dağılıma dönüşüm yapılabilmesi için değişkenin Normal Dağılımlı olması gereklidir.  
* Normal Dağılımlı olmayan değişkenlere de Standardizasyon uygulanabilir ama büyük ihtimal çok efektif bir sonuç alınmayacaktır.  
* Linear Regresyon, Logistic Regresyon gibi dağılımın önemli olduğu ve normallik varsayımlarının bulunduğu algoritmalarda kullanılması önerilir.  
* Standardization değişkenin dağılımını şekil olarak değiştirmez.  
* <span class="mark">Outlier'ların varlığından Nromalization'a göre çok daha az etkilenir. Çünkü bu teknikte değerleri belli bir aralığa sınırlandırmaya gerek yoktur.</span>  
* <span style='color:Blue' >**0-1 şeklindeki nominal kategorik verileri Standardization işlemine sakın sokma!**</span>

**ROBUST SCALER;**  
* Adından da anlaşılacağı üzere outlier'lara karşı daha robust olan bir scaling tekniğidir.  
* Outlier'lar skewed datalarda olur. Yani Normal Dağılımlı olmayan verilerde olurlar. Normal Dağılımlı olmayan verilerde Normalization kullanmayı önerdik ama aykırı değerler çok baskın gelirse Normalization etkilenir dedik. İşte bu durumda Robust Scaler bize bir scaling alternatifi olabilir.  
* Ölçeklendirme IQR kullanılarak yapılır böylece Outlier'ların etkisi azaltılır.  
* Feature Scaling gerektiren her durumda kullanabilirsin.  
* **0-1 şeklindeki nominal kategorik verileri bu işleme dahil etmemen önerilir!**

<span class="burk">**BENİM İZLEDİĞİM YOL;**</span>  
1. Ben değişkenlerin hem raw hem normalization yapılmış hem standardization yapılmış hem de Robust Scaler kullanılmış halleriyle ayrı ayrı modeller kurup performans metriklerini karşılaştırıyorum.  


2. Standardizasyon yapacaksam verileri Normal Dağılımlı yapmak için transformasyonlar da kullanıyorum. 


3. Outlier'ları dikkate alıyorum. Ya onları baskılıyorum ya da Robust Scaler yönetimini tercih ediyorum.

### Scaling Yöntemleri
#### Normalization
* MinMax scaler olarak da bilinir.  
* Değişkenin değerlerini [0,1] aralığına ölçeklendirir. Yani değerler belirli bir aralığa scale edilir.  
* Normal dağılımlı olmayan değişkenlerde kullanılması önerilir.  
* Neural Network ve distance based (KNN...) algoritmalarda kullanılması önerilir.  
* Bu teknik outlier'lara karşı oldukça hassastır, etkilenir. Çünkü değerleri belli bir aralığa sığdırmaya çalışıyoruz ve outlier'lar diğer değerleri baskılayabilir.  
* Normalization değişkenin dağılımında şekil olarak bir değişiklik oluşturmaz.  
* 0-1 şeklindeki nominal kategorik verileri Normalization işlemine sokabilirsin. Aslında kategorik verilerde scaling işlemi kullanılmaz ama Normalization nominal kategorik verilerde hiçbir değişiklik yapamayacağı için onları da işleme dahil edebilirsin. Bir sıkıntı olmaz.  

```python
# Önce verimizi çağıralım;
from sklearn.model_selection import train_test_split
df = pd.read_csv("diamonds.csv")
df1 = df.copy()

# Veriseti X=bağımsız değişkenler ve Y=output olarak ayrılır.
# Ayrıca verisetinden sadece continous değişkenleri almak için kategorik/discrete değişkenleri atalım.
X = df.drop(["price","clarity","color","index","cut"],axis=1)
y = df.price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True, stratify=y, random_state=2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle=True, stratify=y_train,
                                                  random_state=2)

# Shuffle özelliğini açtığımız için index'ler karışmış halde olacaktır. Indexleri düzene sokmak için şu kodları yaz;
for i in [X_train, X_val, X_test, y_train, y_val, y_test]:
    i.reset_index(inplace=True, drop=True)
```

```python
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler

# Fit işlemi Train verisine göre yapılır. Validation/Test setlerinde de Train verisiyle fit edilmiş scaling nesnesi kullanılır.
norm = MinMaxScaler().fit(X_train)

# transform train data
X_train_norm = pd.DataFrame(norm.transform(X_train), columns=X_train.columns)

# transform validation data
X_val_norm = pd.DataFrame(norm.transform(X_val), columns=X_val.columns)

# transform test data
X_test_norm = pd.DataFrame(norm.transform(X_test), columns=X_test.columns)
```

```python
X_train_norm.describe()
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
sns.kdeplot(X_train.x, ax=axes[0])
sns.kdeplot(X_train_norm.x, ax=axes[1]);
# Soldaki Feature Scaling Öncesi, sağdaki Feature Scaling sonrası için çizdirilen grafiklerdir.
# Görüldüğü üzere Normalization dağılımda bir değişiklik oluşturmaz.
```

#### Standardization
* Z-Score Normalization olarak da bilinir.  


* Değişkenleri Standard Normal Dağılıma dönüştürür.  


* Yani; değerleri 0 noktasında merkezleyen, ortalamalarını 0 ve standart sapmalarını 1 yapan bir scaling tekniğidir.  


* Bu tekniğin Normal Dağılımlı değişkenlerle kullanılması önerilir. Çünkü Standard Normal Dağılıma dönüşüm yapılabilmesi için değişkenin Normal Dağılımlı olması gereklidir.  


* Normal Dağılımlı olmayan değişkenlere de Standardizasyon uygulanabilir ama büyük ihtimal çok efektif bir sonuç alınmayacaktır.  


* Linear Regresyon, Logistic Regresyon gibi dağılımın önemli olduğu ve normallik varsayımlarının bulunduğu algoritmalarda kullanılması önerilir.  


* Outlier'ların varlığından Nromalization'a göre çok daha az etkilenir. Çünkü bu teknikte değerleri belli bir aralığa sınırlandırmaya gerek yoktur.  


* Standardization değişkenin dağılımını şekil olarak değiştirmez.  


* **0-1 şeklindeki nominal kategorik verileri Standardization işlemine sakın sokma!**

```python
# Önce verimizi çağıralım;
from sklearn.model_selection import train_test_split
df = pd.read_csv("diamonds.csv")
df1 = df.copy()

# Veriseti X=bağımsız değişkenler ve Y=output olarak ayrılır.
# Ayrıca verisetinden sadece continous değişkenleri almak için kategorik/discrete değişkenleri atalım.
X = df.drop(["price","clarity","color","index","cut"],axis=1)
y = df.price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True, stratify=y, random_state=2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle=True, stratify=y_train,
                                                  random_state=2)

# Shuffle özelliğini açtığımız için index'ler karışmış halde olacaktır. Indexleri düzene sokmak için şu kodları yaz;
for i in [X_train, X_val, X_test, y_train, y_val, y_test]:
    i.reset_index(inplace=True, drop=True)
```

```python
# data standardization with sklearn
from sklearn.preprocessing import StandardScaler

# Fit işlemi Train verisine göre yapılır. Validation/Test setlerinde de Train verisiyle fit edilmiş scaling nesnesi kullanılır.
scaler = StandardScaler().fit(X_train)

# transform train data
X_train_stnd = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)

# transform validation data
X_val_stnd = pd.DataFrame(scaler.transform(X_val), columns=X_val.columns)

# transform test data
X_test_stnd = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
X_train_stnd.describe()
# Görüldüğü üzere mean=0 ve std=1 olmuş tüm değişkenlerde.
# Ama bu yöntemin etkin olabilmesi için değişkenlerin Normal Dağılımlı olması önerilir.
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
sns.kdeplot(X_train.x, ax=axes[0])
sns.kdeplot(X_train_stnd.x, ax=axes[1]);
# Soldaki Feature Scaling Öncesi, sağdaki Feature Scaling sonrası için çizdirilen grafiklerdir.
# Görüldüğü üzere Standardization dağılımda bir değişiklik oluşturmaz.
```

#### Robust Scaler
* Adından da anlaşılacağı üzere outlier'lara karşı daha robust olan bir scaling tekniğidir.  


* Outlier'lar skewed datalarda olur. Yani Normal Dağılımlı olmayan verilerde olurlar. Normal Dağılımlı olmayan verilerde Normalization kullanmayı önerdik ama aykırı değerler çok baskın gelirse Normalization etkilenir dedik. İşte bu durumda Robust Scaler bize bir scaling alternatifi olabilir.  


* Ölçeklendirme IQR kullanılarak yapılır böylece Outlier'ların etkisi azaltılır.  


* Feature Scaling gerektiren her durumda kullanabilirsin.  


* **0-1 şeklindeki nominal kategorik verileri bu işleme dahil etmemen önerilir!**

## SMOTE (for Imbalanced Dataset in Classification)
* Synthetic Minority Oversampling TEchnique


* Classification problemlerinde imbalanced bir hedef değişkenle karşı karşıya kalındığında kullanılan etkili bir tekniktir/algoritmadır.  


* Bu teknik genel anlamıyla hedef değişkende daha az sayıdaki sınıfın sayısının arttırılmasına dayanır. Bunun için verisetinde mantıklı bir şekilde sentetik gözlemler üretilir.  


* Çalışma prensibi; algoritma hedef değişkendeki minör sınıftan random bir gözlem seçiyor ve bunun "k" sayıdaki en yakın komşusundan da random bir gözlem seçiyor ve bu iki gözlem arasında bir bağlantı kurup(muhtemelen ortalamalarını alıyordur) yeni bir sentetik gözlem oluşturuyor. Yani arka planda K-NN algoritması gibi çalışıyor diyebiliriz.  


* SMOTE tekniği sadece Train seti üzerinde uygulanır. Validation ve Test setlerinde uygulanmaz.


* SMOTE tekniği kesinlikle data splitten ve tüm feature processing/engineerin aşamaları tamamlandıktan sonra yapılmalıdır.  


* Ayrıca kullanılan fonksiyon tüm değerlerin nümerik olmasını istiyor. O nedenle tüm kategorik değerleri encode etmiş olmalısın.  

<span class="mark">**UYGULAMA**</span>
* 2 tane önemli parametre var ve bunlar düzenlenerek SMOTE uygulaması etkin şekilde yapılabilir.


* sampling_strategy; Hedef değişkendeki minor sınıfın sayısının major sınıfın yüzde kaçına kadar arttırılacağını belirtir. Eğer 0.5 girilirse minor sınıfın sayısı major sınıfın yarısına eşit olana kadar arttırılır. Default=1, yani minor ve major sınıfların sayısını eşitler.


* k_neighbours; algoritmanın sentetik veri oluştururken kaç adet en yakın komşu gözleme bakması gerektiğini belirtir.


* ML projelerinde SMOTE işlemlerinde bu bahsettiğimiz parametrelerde farklı farklı değerler uygulanarak modelin tahmin gücüne en optimum etki yapanın kullanılmasına çalışılır. Yani birden fazla varyasyonu deneyerek en iyisini bulmaya çalışacağız.

```python
#!pip install imbalanced-learn
from imblearn.over_sampling import SMOTE
```

```python
# Önce verinin import edilmesi ve split edilmesi tamamlanır.
df = pd.read_csv("train.csv")
df1 = df.copy()

# Eksik verileri örneğimiz için rastgele dolduralım. Ayrıca zaman kaybı olmasın diye kategorik verileri direkt attık.
# Çünkü SMOTE fonksiyonu tüm değişkenlerin nümerik olmasını istiyor. Kategorikleri encode etmekle uğraşmayalım, zaman kaybı olur
df.Age.fillna(30, inplace=True)
df.Embarked.fillna("S", inplace=True)
df.drop(["Cabin","Name","Embarked","Sex","Ticket"], axis=1, inplace=True)

# Bağımlı ve bağımsız değişkenleri ayıralım.
X = df.drop(["Survived"],axis=1)
y = df.Survived

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True, random_state=2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, shuffle=True, random_state=2)
```

```python
#SMOTE işlemi;
oversample = SMOTE(sampling_strategy=0.6, k_neighbors=7, random_state=2)
X_train_sm, y_train_sm = oversample.fit_resample(X_train, y_train)
```

```python
print("SMOTE öncesi:\n",y_train.value_counts())
```

```python
print("SMOTE sonrası:\n",y_train_sm.value_counts())
# SMOTE fonksiyonunda da belirtildiği üzere minor sınıfın sayısı major sınıfın %60'ına eşitlene kadar sentetik veri üretildi.
```
